```python
import tensorflow as tf
from keras.applications.vgg16 import VGG16
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Inisialisasi model VGG19
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

```python
# Membekukan lapisan-lapisan awal (hingga Conv5) dari model VGG19
for layer in base_model.layers:
    layer.trainable = False
```

```python
# Membangun model klasifikasi di atas VGG19
x = Flatten()(base_model.output)
output = Dense(2, activation='softmax')(x)  
model = Model(inputs=base_model.input, outputs=output)
```

```python
# Kompilasi model
model.compile(loss='categorical_crossentropy',
              optimizer=Adam(learning_rate=0.0001),
              metrics=['accuracy'])
```

```python
# Preprocessing data dan augmentasi gambar
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')

test_datagen = ImageDataGenerator(rescale=1.0 / 255)
```

```python
# Persiapkan data pelatihan dan validasi
train_generator = train_datagen.flow_from_directory(
    'C:/Users/GF63/train_data/train',  # Gantilah 'train_dir' dengan direktori data latih Anda
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical')

validation_generator = test_datagen.flow_from_directory(
    'C:/Users/GF63/train_data/val',  # Gantilah 'validation_dir' dengan direktori data validasi Anda
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical')
```

```python
# Pelatihan model
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // 32,
    epochs=10,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // 32,
    verbose=1)
```

```python
# loss
import matplotlib.pyplot as plt
plt.plot(history.history['loss'], label='train loss')
plt.plot(history.history['val_loss'], label='val loss')
plt.legend()
plt.show()
```

```python
plt.plot(history.history['accuracy'], label='train acc')
plt.plot(history.history['val_accuracy'], label='val acc')
plt.legend()
plt.show()
```

```python
# Simpan model ke file
model.save('vgg19_v2.h5')
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import load_model
import os
```

```python
model = load_model(os.path.join('vgg19_v2.h5'))
```

```python
# Fungsi untuk menampilkan confusion matrix
def plot_confusion_matrix(y_true, y_pred, classes):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.set()
    sns.heatmap(cm, annot=True, fmt='d', cmap='Greys', xticklabels=classes, yticklabels=classes)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    plt.show()
```

```python
# Prediksi menggunakan model
test_datagen = ImageDataGenerator(rescale=1.0 / 255)
test_generator = test_datagen.flow_from_directory(
    'C:/Users/GF63/train_data/test',  # Gantilah 'test_dir' dengan direktori data uji Anda
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=False)  # Pastikan tidak mengocok urutan data uji
```

```python
y_true = test_generator.classes  # Kelas sebenarnya dari data uji
y_pred = np.argmax(model.predict(test_generator), axis=-1)  # Prediksi model
```

```python
# Tampilkan confusion matrix
class_names = list(test_generator.class_indices.keys())
plot_confusion_matrix(y_true, y_pred, classes=class_names)
```

```python
# Tampilkan laporan klasifikasi
print(classification_report(y_true, y_pred, target_names=class_names))
```
